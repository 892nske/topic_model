```python
import MeCab
```

```python
tagger = MeCab.Tagger()
text = "彼は３浪までして念願の東京大学法学部に合格できました"
```

```python
node = tagger.parseToNode(text)
while node:
    print("単語の表記：",node.surface)
    print("単語の属性：",node.feature)
    node_features = node.feature.split(",")
    print("単語の属性（リスト）：",node_features)
    node = node.next
```
